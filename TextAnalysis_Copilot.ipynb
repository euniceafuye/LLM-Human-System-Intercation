```python

def create_wordcloud(text, title, colormap='viridis'):
    wordcloud = WordCloud(
        width=800,
        height=400,
        background_color='white',
        colormap=colormap,
        max_words=30,
        contour_width=1,
        contour_color='steelblue'
    ).generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, fontsize=16, pad=20)
    plt.show()

create_wordcloud(text_copilot_full, "Copilot - Key Words", 'Greens')
```

## Sentiment Analysis

```python
def sentiment_by_output(outputs, label="Gemini"):
    print(f"\nSENTIMENT ANALYSIS – {label}")
    print("="*50)
    polarities = []
    subjectivities = []

    for i, risposta in enumerate(outputs, 1):
        blob = TextBlob(risposta)
        p = blob.sentiment.polarity
        s = blob.sentiment.subjectivity

        polarities.append(p)
        subjectivities.append(s)

        if p > 0.1:
            tono = "POSITIVE"
        elif p < -0.1:
            tono = "NEGATIVE"
        else:
            tono = "NEUTRAL"

        print(f"\nResponse {i}:")
        print(f"  Polarity:      {p:.3f} ({tono})")
        print(f"  Subjectivity:  {s:.3f}")

    # Analisi complessiva
    full_text = " ".join(outputs)
    blob_full = TextBlob(full_text)
    p_full = blob_full.sentiment.polarity
    s_full = blob_full.sentiment.subjectivity

    print("\nOverall:")
    print(f"  Global polarity:     {p_full:.3f}")
    print(f"  Global subjectivity: {s_full:.3f}")

    # Grafico riassuntivo
    labels = ['Polarity', 'Subjectivity']
    values = [p_full, s_full]
    colors = ['lightgreen' if v >= 0 else 'lightcoral' for v in values]

    plt.figure(figsize=(6, 5))
    bars = plt.bar(labels, values, color=colors, edgecolor='black', width=0.5)
    plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
    plt.ylabel('Value (-1 to +1)')
    plt.title(f'Overall Sentiment – {label}')
    plt.ylim(-1, 1)
    plt.grid(axis='y', alpha=0.3)

    for bar, v in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width()/2,
                 v + 0.05 if v >= 0 else v - 0.1,
                 f'{v:.3f}',
                 ha='center',
                 fontweight='bold')

    plt.tight_layout()
    plt.show()

    return polarities, subjectivities, p_full, s_full

pol_list_g, subj_list_g, p_g_full, s_g_full = sentiment_by_output(copilot_outputs, label="Copilot")
```

```python
import re
from collections import Counter

# Unisci tutte le risposte in un unico testo
text_copilot_full = " ".join(copilot_outputs).lower()

mentor_words = [
    "mentor", "guidance", "guiding", "support", "supportive",
    "encouragement", "encouraging", "feedback", "refine", "refinement"
]

academic_words = [
    "validity", "epistemic", "rigor", "rigorous", "bias", "sycophancy",
    "automation", "grounding", "pedagogical", "metric", "scale",
    "standardisation", "reproducibility", "cognitive", "offloading"
]

def count_occurrences(words, text):
    counts = {}
    for w in words:
        counts[w] = len(re.findall(rf"\b{re.escape(w)}\b", text))
    return counts

mentor_counts   = count_occurrences(mentor_words, text_copilot_full)
academic_counts = count_occurrences(academic_words, text_copilot_full)

print("MENTOR-LIKE WORDS")
for w, c in mentor_counts.items():
    print(f"{w}: {c}")

print("\nACADEMIC WORDS")
for w, c in academic_counts.items():
    print(f"{w}: {c}")
```
